```coconut
def fizzbuzz(n):
    case n:
        match n if (n % 3) == 0 and (n % 5) == 0:
            print("FizzBuzz")
        match n if (n % 3) == 0:
            print("Fizz")
        match n if (n % 5) == 0:
            print("Buzz")
        match n:
            print(n)

range(1, 101) |> map$(fizzbuzz) |> list
```
